# Cagetti & De Nardi (2006) 

This notebook will explore and replicate the life-cycle occupational-choice model of [Cagetti and De Nardi (2006)](https://www.journals.uchicago.edu/doi/10.1086/508032). The main goal of the paper is to understand how entrepreneurial frictions affect saving behavior and the distribution of wealth. Cagetti and De Nardi build a parsimonious life-cycle model in which households choose between working for wages and operating a business. Entrepreneurs face endogenous borrowing constraints that arise from limited contract enforceability, meaning lenders restrict borrowing because repayment cannot be perfectly enforced. These financial frictions influence entrepreneurial saving behavior, firm size, aggregate capital accumulation, and ultimately help explain the concentration of wealth observed in the data.

Within this notebook, I will first introduce their model using the notation in the paper. I will then move to the replication exercise, but apply different solving method compare with the paper. 



## The Model

### Demographics / household objective

The model is a life-cycle model with intergenerational altruism. Agents are either young or old. A young person becomes old with probability $1-\pi_y$, while an old person survives as old with probability $\pi_o$; when an old person dies, a descendant is born. The household's objective is to maximize expected discounted lifetime utility from consumption, with CRRA period utility $u(c) = c^{1-\sigma}/(1-\sigma)$, discount factor $\beta$, and altruistic weight $\eta$ on descendants' utility. When $\eta=0$, the household behaves like a pure life-cycle agent; when $\eta=1$, it behaves like a dynasty.

So the objective is not firm value. It is household lifetime utility, where entrepreneurship is one possible occupational/investment choice inside that dynamic problem.

### Abilities, technology, and borrowing frictions

At the core of the model, households differ in two abilities:

- $y$: worker ability (capacity to produce income out of labor)
- $\theta$: entrepreneurial ability (capacity to invest capital more or less productively)

Entrepreneurs can operate a technology with decreasing returns: when the entrepreneur invests $k$, production is $\theta k^{\nu}$ with $\nu \in [0,1]$. They can borrow to finance capital $k$, but because contracts are imperfectly enforceable, borrowing is limited by an endogenous no-default condition: if an entrepreneur defaults, he can abscond with a fraction $f$ of working capital and work as a worker for one period. This is what makes wealth matter as collateral in the model.

Workers can save (but not borrow) at the risk-free rate $r$. The nonentrepreneurial sector is represented by a standard Cobb–Douglas production function $F(K_c, L_c) = A K_c^{\alpha} L_c^{1-\alpha}$. In both sectors, capital depreciates at rate $\delta$.

### Households

#### State variables and timing

At the beginning of each period, before any decisions are made, current ability levels are known with certainty while next period's levels are uncertain. For a young agent the state is $(a, y, \theta)$: current assets $a$, worker ability $y$, and entrepreneurial ability $\theta$. For an old entrepreneur the state is summarized by $W(a, \theta)$—current assets, entrepreneurial ability, and (implicitly) whether he was an entrepreneur or retiree last period—with the option to continue or retire. In the equilibrium section the paper gives the full individual state vector as $\boldsymbol{x} = (a, y, \theta, s)$, where $s$ indicates young worker, young entrepreneur, old entrepreneur, or old retired.

#### Controls

The household makes a discrete choice (occupation when young, continue vs. retire when old) and then continuous choices. A young person first chooses occupation—worker or entrepreneur. If entrepreneur: consumption $c$, business capital $k$, and next-period assets $a'$. If worker: $c$ and $a'$. An old entrepreneur first chooses whether to stay in business or retire; if he continues, he chooses $c$, $k$, and $a'$; if retired, he chooses $c$ and $a'$.

#### Shocks and timing within the period

The only exogenous stochastic variables are next period's ability draws—worker ability $y'$ and entrepreneurial ability $\theta'$. Within a period: the household observes $(a_t, y_t, \theta_t)$, chooses occupation/status and continuous controls, and enjoys consumption at the end of the period; next-period abilities $(y_{t+1}, \theta_{t+1})$ are uncertain and enter through expectations. For the young, the Bellman expectation is over $(y', \theta')$ conditional on $(y, \theta)$, with $F(y', \theta' \mid y, \theta)$ first-order Markov. For old entrepreneurs, the child's value is integrated over the child's abilities: $y'$ from the invariant distribution and $\theta'$ conditional on the parent's $\theta$ (reflecting firm inheritance).

## Bellman equations

The value functions below are as in the paper. The young choose between entrepreneurship and work; the old choose between continuing the business and retiring.

### The young's problem

The young household's value function is

$$V(a, y, \theta) = \max \left\{ V_e(a, y, \theta),\, V_w(a, y, \theta) \right\}. \tag{2}$$

If the young chooses to be an entrepreneur,

$$V_e(a, y, \theta) = \max_{c, k, a'} \left\{ u(c) + \beta \pi_y E V(a', y', \theta') + \beta (1-\pi_y) E W(a', \theta') \right\}, \tag{3}$$

subject to

$$a' = (1-\delta) k + \theta k^{\nu} - (1+r)(k-a) - c, \tag{4}$$

$$u(c) + \beta \pi_y E V(a', y', \theta') + \beta (1-\pi_y) E W(a', \theta') \geq V_w(f \cdot k, y, \theta), \tag{5}$$

$$a \geq 0, \tag{6}$$

$$k \geq 0. \tag{7}$$

Equation (5) is the endogenous borrowing/enforcement constraint.

If the young chooses to be a worker,

$$V_w(a, y, \theta) = \max_{c, a'} \left\{ u(c) + \beta \pi_y E V(a', y', \theta') + \beta (1-\pi_y) W_r(a') \right\}, \tag{8}$$

subject to

$$a' = (1+r) a + (1-\tau) w y - c. \tag{9}$$

($w$ is the wage and $\tau$ is the proportional payroll tax financing old-age social security.)

### The old's problem

An old entrepreneur chooses whether to stay in business or retire. His value function is

$$W(a, \theta) = \max \left\{ W_e(a, \theta),\, W_r(a) \right\}. \tag{10}$$

If the old entrepreneur stays in business,

$$W_e(a, \theta) = \max_{c, k, a'} \left\{ u(c) + \beta \pi_o E W(a', \theta') + \eta \beta (1-\pi_o) E V(a', y', \theta') \right\}, \tag{11}$$

subject to (4), (6), and (7), and

$$u(c) + \beta \pi_o E W(a', \theta') + \eta \beta (1-\pi_o) E V(a', y', \theta') \geq W_r(f \cdot k). \tag{12}$$

If the old person is retired,

$$W_r(a) = \max_{c, a'} \left\{ u(c) + \beta \pi_o W_r(a') + \eta \beta (1-\pi_o) E V(a', y', \theta') \right\}, \tag{13}$$

subject to

$$a' = (1+r) a + p - c. \tag{14}$$

($p$ is pensions and social security payments.)

In the paper, the authors calibrate the model by selecting parameter values based on empirical evidence and then using these values to generate quantitative results. In this notebook, however, I plan to solve the Bellman equation using DSOP methods, which focus on the structure of the optimization problem rather than on numerical calibration. As a result, the solution does not require inserting specific parameter values.

The model should not be treated as a single within-period optimization problem because its decisions occur sequentially and depend on earlier choices. At the beginning of a period, the household observes its current assets and ability. A young agent first chooses whether to become a worker or an entrepreneur, while an old entrepreneur chooses whether to continue operating the business or retire. This discrete decision determines which continuous optimization problem becomes relevant. Workers and retirees choose consumption and next-period assets, whereas entrepreneurs must also choose business capital and satisfy the endogenous no-default constraint.

Thus, one decision changes the feasible set, budget constraint, and continuation value of the next decision. Future ability shocks are not known when current decisions are made and therefore enter the Bellman equation only through expectations. Decomposing the problem into stages, as in DSOP/DDSL, makes computation easier, simulation clearer, and the economic mechanisms more transparent by separating occupational choice, saving behavior, entrepreneurial scale choice, and borrowing constraints.




## Stage decomposition

Here are several problems:

1. Before these four pipelines, there is a `branch` stage in which the young household chooses its occupation, and the old entrepreneur decides whether to continue the business or retire.
2. The shock occurs before the pipelines and affects the next-period ability $(y',\theta')$.

### Young worker
`cons-only` type period

| Stage | Role | Arrival perch | Control | Continuation perch |
|-------|------|---------------|---------|--------------------|
| `portable(0)` | Build cash-on-hand from assets and labor income | $(a, y, \theta)$ | none | $m^w$ |
| `cons-noshocks` | Choose consumption/saving | $m^w$ | $c$ | $a'$ |
| `disc` | Apply preference discounting to continuation object | continuation values from next period | none | $V_w(a, y, \theta)$ |

$m^w = (1+r)a + (1-\tau) w y, \qquad a' = m^w - c.$

Define the worker post-decision continuation:

$\bar{V}_w(a'; y, \theta) = \beta \pi_y E[V(a', y', \theta')] + \beta(1-\pi_y) W_r(a')$. 

Then $V_w(a, y, \theta) = \max_{a' \geq 0} \left\{ u\big( (1+r)a + (1-\tau)wy - a' \big) + \bar{V}_w(a'; y, \theta) \right\}.$

### Young entrepreneur
`port-cons` type period
| Stage | Role | Arrival perch | Control | Continuation perch |
|-------|------|---------------|---------|--------------------|
| `portable(optimize k)` | Choose entrepreneurial scale and build pre-consumption resources | $(a, y, \theta)$ | $k$ | $m^e$ |
| `cons-noshocks` | Choose consumption/saving after scale is set | $m^e$ | $c$ | $a'$ |
| `disc` | Apply preference discounting to continuation object | continuation values from next period | none | $V_e(a, y, \theta)$ |

$m^e = (1-\delta)k + \theta k^\nu - (1+r)(k-a), \quad a' = m^e - c.$

Define the entrepreneur post-decision continuation:

$\bar{V}_e(a'; y, \theta) = \beta \pi_y E[V(a', y', \theta')] + \beta(1-\pi_y) E[W(a', \theta')]$. 

Then $V_e(a, y, \theta) = \max_{k \geq 0,\, a' \geq 0} \left\{ u\big( (1-\delta)k + \theta k^\nu - (1+r)(k-a) - a' \big) + \bar{V}_e(a'; y, \theta) \right\}$.

### Old continuing entrepreneur
`port-cons` type period

| Stage | Role | Arrival perch | Control | Continuation perch |
|-------|------|---------------|---------|--------------------|
| `portable(optimize k)` | Choose business scale and build pre-consumption resources | $(a, \theta)$ | $k$ | $m^e$ |
| `cons-noshocks` | Choose consumption/saving after scale is set | $m^e$ | $c$ | $a'$ |
| `disc` | Apply preference discounting and altruism weighting $\eta$ | continuation values from next period | none | $W_e(a, \theta)$ |

$m^e = (1-\delta)k + \theta k^\nu - (1+r)(k-a), \quad a' = m^e - c.$

Define the continuing-entrepreneur post-decision continuation:

 $\bar{W}_e(a'; \theta) = \beta \pi_o E[W(a', \theta')] + \eta \beta(1-\pi_o) E[V(a', y', \theta')]$.
 
Then $W_e(a, \theta) = \max_{k \geq 0,\, a' \geq 0} \left\{ u\big( (1-\delta)k + \theta k^\nu - (1+r)(k-a) - a' \big) + \bar{W}_e(a'; \theta) \right\}$.

### Old retiree
`cons-only` type period

| Stage | Role | Arrival perch | Control | Continuation perch |
|-------|------|---------------|---------|--------------------|
| `portable(0)` | Build cash-on-hand from assets and pension income | $a$ | none | $m^r$ |
| `cons-noshocks` | Choose consumption/saving | $m^r$ | $c$ | $a'$ |
| `disc` | Apply preference discounting and altruism weighting $\eta$ | continuation values from next period | none | $W_r(a)$ |

$m^r = (1+r)a + p, \qquad a' = m^r - c.$

Define the retiree post-decision continuation:

$\bar{W}_r(a') = \beta \pi_o W_r(a') + \eta \beta(1-\pi_o) E[V(a', y', \theta')]$. 

Then $W_r(a) = \max_{a' \geq 0} \left\{ u\big( (1+r)a + p - a' \big) + \bar{W}_r(a') \right\}.$

### Connector between periods

In all four pipelines, the only continuous state that carries across time is assets. The connector takes the post-decision asset level $a'$ from period $t$'s continuation perch and passes it as next period's beginning-of-period assets $a$ at the arrival perch of period $t+1$. So the connector is $a' \mapsto a$.


## Comparison with the monolithic formulation

The staged decomposition does not change the model: it is a different way of reading the same Bellman equations (2)–(14). The top-level max in (2) and (10) is the discrete stage (occupation / continue vs retire); the inner maxes in (3), (8), (11), and (13) are the continuous stage on each branch; expectations over $(y',\theta')$ and discounting sit in the continuation objects and the connector. The decomposition makes three things explicit: the order of decisions (discrete then continuous, with branch-specific constraints), the fact that only $a'$ carries to the next period (entrepreneurial $k$ is within-period only), and where uncertainty enters (at next period’s arrival, via the connector). Other orderings are possible, but this one matches the paper’s “observe state → choose branch → choose $(c,k,a')$ on that branch,” so the four pipelines and the connector map directly onto the four value functions and their expectations. The paper already implies this structure; I only make the stages and connector explicit.
